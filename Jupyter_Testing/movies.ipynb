Calculate Weighted Movie Rating



WR = (v Ã· (v+m)) Ã— R + (m Ã· (v+m)) Ã— C



v= number of votes for a movie



m = minimum number of votes required



R = average rating of the movie



C = average rating accross all movies





```python
m = movies_df['vote_count'].quantile(0.9)

C = movies_df["vote_average"].mean()
```

```python
def weighted_rating(df, m=m, C=C):
    R = df["vote_average"]
    v = df["vote_count"]
    wr = ((v / v+m) * R) + (m / (v+m) * C)
    return wr
```

```python
movies_df_v2 = movies_df.copy().loc[movies_df["vote_count"] >= m]

movies_df_v2["weighted_rating"] = movies_df_v2.apply(weighted_rating, axis=1)
movies_df_v2["rank"] = movies_df_v2["weighted_rating"].rank(ascending=False)
movies_df_v2 = movies_df_v2.sort_values(by="rank", ascending=True)


movies_df_v3 = movies_df_v2[["title", "release_date", "tagline", "runtime", "original_language", "vote_average", "vote_count", "rank"]]

movies_df_v3["runtime"] = movies_df_v3["runtime"].divide(60).round(2)

movies_top_50_df = movies_df_v3.sort_values(by="rank").head(50)
movies_top_50_df
```
